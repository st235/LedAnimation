```python
def create_image_from_color(color):
    image = np.zeros((32, 32, 3))
    image[:, :] = [i / 255.0 for i in color]
    return image

cols = 10
rows = len(colors) // cols + 1 if len(colors) % cols > 0 else 0

fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(20, 20))

for index, color in enumerate(colors):
        i = index // cols
        j = index % cols

        ax[i, j].set_title(color[0])
        ax[i, j].imshow(create_image_from_color(color[1:]))

for i in range(rows):
  for j in range(cols):
    ax[i,j].axis("off")


plt.tight_layout()
plt.show()
fig.savefig('palette.png')
```
